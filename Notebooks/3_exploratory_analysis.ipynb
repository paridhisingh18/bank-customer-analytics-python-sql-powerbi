#### UNIVARIATE ANALYSIS: Categorical Variables Distribution

```python
# Examining the distribution of unique categories in categorical columns
categorical_cols = df[['Gender', 'Banking Relationship','Investment Advisor','Nationality','Amount of Credit Cards','Occupation', 'Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band','Age','Age Groups']]

for col in categorical_cols:
    print(f"value counts for '{col}' :")
    display(df[col].value_counts())
```

#### VISUALIZATION: Categorical Variable Count Plots

```python
for i, predictor in enumerate(categorical_cols):
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=predictor)
    plt.title(f"Distribution of {predictor}")
    plt.xticks(rotation=90)
    plt.show()
```

##### UNIVARIATE ANALYSIS: Numerical Variables Distribution

```python
numerical_cols = df[['Age','Estimated Income','Estimated Annual Income','Net Worth','Debt to Income Ratio','Property Value','Bank Loans', 'Bank Deposits']]

for col in numerical_cols:
    plt.figure(figsize = (8,5))
    sns.histplot(data = df , x= col , kde= True  , bins  = 20 , edgecolor = 'black', alpha = 0.6)

    mean_value = df[col].mean()
    median_value = df[col].median()

    plt.axvline(mean_value , color = 'Red' , linestyle= '--' , linewidth=2, label=f'Mean: {mean_value:.2f}')
    plt.axvline(median_value , color = 'Green' , linestyle= '--' , linewidth=2, label=f'Mean: {mean_value:.2f}')


    plt.title(f"{col} distribution")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()
```

### ðŸ“Š Insight 

The analysis of financial variables such as **Income**, **Deposits**, **Loans**, and **Debt-to-Income Ratio** indicates a predominantly **middle-income customer base**.  
However, a small segment of **high-value clients** has a **disproportionate impact** on the overall financial metrics, resulting in **right-skewed distributions** across these features.


##### SEGMENT ANALYSIS: Age Group Performance

```python
# Age Group Vs Net worth and Income 

age_summary = df.groupby('Age Groups')[['Estimated Annual Income' , 'Net Worth']].median().reset_index()

metrics = ['Estimated Annual Income' , 'Net Worth']

for metric in metrics:
    plt.figure(figsize = (8,5))
    sns.barplot(x= 'Age Groups' , y= metric , data = age_summary )
    plt.title(f"{metric} by age group")
    plt.xlabel("Age Groups")
    plt.ylabel(metric)
    plt.tight_layout()
    plt.show()

    
```

```python
loyalty_dti = df.groupby('Loyalty Classification')['Debt to Income Ratio'].mean().reset_index()

plt.figure(figsize = (8,6))
sns.barplot(x = 'Loyalty Classification', y = 'Debt to Income Ratio' , data = loyalty_dti)
plt.title("LOYALTY TIER ANALYSIS")
plt.xlabel("Loyalty Classification")
plt.ylabel("Debt to Income Ratio")
plt.tight_layout()
plt.show()
```

- Loyal customers generally maintain lower debt-to-income ratios, indicating stronger financial stability and lower credit risk.

```python
gender_summary = df.groupby("Gender")[['Estimated Annual Income' , 'Net Worth']].mean().reset_index()

metrics = ["Estimated Annual Income" , "Net Worth"]

for metric in metrics:
    plt.figure(figsize=(6,4))
    sns.barplot(x="Gender", y=metric, data=gender_summary)
    plt.title(f"{metric} by Gender")
    plt.xlabel("Gender")
    plt.ylabel(metric)
    plt.tight_layout()
    plt.show()


```

#### Correlation Matrix

```python
# Select numerical columns for correlation analysis
numerical_cols = ['Age', 'Estimated Income', 'Estimated Annual Income','Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned','Net Worth','Debt to Income Ratio']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

### Correlation Heatmap â€” Key Insights
**Strongest Correlation:**  
  - **Net Worth â†” Properties Owned (0.99)** â€” Higher property ownership strongly drives overall net worth.

**Financial Behavior Patterns:**  
  - **Bank Deposits â†” Checking Accounts (0.84)** and **Savings â†” Checking (0.75)** â€” Indicate strong liquidity management and disciplined saving habits.

**Income Dynamics:**  
  - **Income â†” Superannuation Savings (0.37)** â€” Higher-income customers invest more in long-term savings.  
  - **Income â†” DTI (-0.45)** â€” Clear inverse relationship; financially stable customers manage debt efficiently.

**Weak or Unrealistic Correlations:**  
  - **Age â‰ˆ 0 with all variables** â€” Suggests a synthetic dataset with limited demographic variation.  
  - **Net Worth â†” DTI (-0.02)** â€” Minimal link between wealth and debt ratio.

**Summary:**  
Wealth-related variables (Net Worth, Deposits, Savings) are strongly interconnected, reflecting sound financial patterns.  
However, negligible age correlations indicate non-realistic demographic behavior, implying a simulated dataset.



```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
    ('Properties Owned', 'Net Worth'),
    ('Age', 'Net Worth')
    
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(6, 6))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},     # semi-transparent points
        line_kws={'color': 'red'}       # best-fit line color
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```
